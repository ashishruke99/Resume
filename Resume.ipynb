```python
import json

# Template for Resume
resume_template = {
    "Personal Information": {
        "Name": "",
        "Contact": {
            "Email": "",
            "Phone": "",
            "Address": ""
        },
        "LinkedIn": "",
        "Additional Info": ""
    },
    "Education": [],
    "Work Experience": [],
    "Skills": [],
    "Certifications": [],
    "Personal Projects": [],
    "Competitions": [],
    "Publications": [],
    "Copyrights": []
}

def create_resume(data):
    resume = resume_template.copy()

    # Personal Information
    resume["Personal Information"]["Name"] = data.get("name", "")
    resume["Personal Information"]["Contact"]["Email"] = data.get("email", "")
    resume["Personal Information"]["Contact"]["Phone"] = data.get("phone", "")
    resume["Personal Information"]["Address"] = data.get("address", "")
    resume["Personal Information"]["LinkedIn"] = data.get("linkedin", "")
    resume["Personal Information"]["Additional Info"] = data.get("additional_info", "")

    # Education
    resume["Education"] = [
        {
            "Degree": edu.get("degree", ""),
            "Institution": edu.get("institution", ""),
            "Duration": edu.get("duration", ""),
            "CGPA": edu.get("cgpa", ""),
            "Location": edu.get("location", "")
        }
        for edu in data.get("education", [])
    ]

    # Work Experience
    resume["Work Experience"] = [
        {
            "Job Title": work.get("job_title", ""),
            "Company": work.get("company", ""),
            "Duration": work.get("duration", ""),
            "Location": work.get("location", ""),
            "Achievements/Tasks": work.get("achievements_tasks", [])
        }
        for work in data.get("work_experience", [])
    ]

    # Skills
    resume["Skills"] = data.get("skills", [])

    # Certifications
    resume["Certifications"] = [
        {
            "Certification": cert.get("certification", ""),
            "Institution": cert.get("institution", ""),
            "Date": cert.get("date", "")
        }
        for cert in data.get("certifications", [])
    ]

    # Personal Projects
    resume["Personal Projects"] = [
        {
            "Project": proj.get("project", ""),
            "Description": proj.get("description", "")
        }
        for proj in data.get("personal_projects", [])
    ]

    # Competitions
    resume["Competitions"] = [
        {
            "Competition": comp.get("competition", ""),
            "Date": comp.get("date", ""),
            "Achievement": comp.get("achievement", "")
        }
        for comp in data.get("competitions", [])
    ]

    # Publications
    resume["Publications"] = [
        {
            "Title": pub.get("title", ""),
            "Publication": pub.get("publication", ""),
            "Date": pub.get("date", "")
        }
        for pub in data.get("publications", [])
    ]

    # Copyrights
    resume["Copyrights"] = [
        {
            "Title": copy.get("title", ""),
            "Registration Number": copy.get("registration_number", ""),
            "Date": copy.get("date", "")
        }
        for copy in data.get("copyrights", [])
    ]

    return resume

# Example data
data = {
    "name": "Ashish Ruke",
    "email": "ashish2002ruke@gmail.com",
    "phone": "N/A",
    "address": "Pune, India",
    "linkedin": "N/A",
    "additional_info": "3 x Hackathon Winner",
    "education": [
        {
            "degree": "Bachelor of Technology - BTech",
            "institution": "Vishwakarma Institute of Information Technology",
            "duration": "01/2022 - Present",
            "cgpa": "8.86",
            "location": "Pune, India"
        },
        # Add more education entries as needed
    ],
    "work_experience": [
        {
            "job_title": "Software and Data Analysis Freelance",
            "company": "SkyGram Stays",
            "duration": "06/2024 - Present",
            "location": "Pune, India",
            "achievements_tasks": [
                "Building a website for booking villa & and analysis customer data"
            ]
        },
        # Add more work experience entries as needed
    ],
    "skills": [
        "Python", "C++", "HTML/CSS", "SQL", "Machine Learning", "LLM", "NLP", "Generative AI", "Agile Scrum",
        "Object-oriented programming", "Debugging", "Deep Learning", "Data Analysis", "Data Visualization",
        "Statistics", "Power BI", "Tableau"
    ],
    "certifications": [
        {
            "certification": "Data Analysis Using Python",
            "institution": "IBM",
            "date": "03/2023"
        },
        # Add more certification entries as needed
    ],
    "personal_projects": [
        {
            "project": "Stock Market Prediction using Deep Learning",
            "description": "Developed web app predicting Nasdaq stock prices for next 30 days using Python, Stacked LSTM model, and Streamlit. Displayed results via interactive graphs"
        },
        # Add more personal projects as needed
    ],
    "competitions": [
        {
            "competition": "Valiant2K23 ML Hackathon",
            "date": "12/2023",
            "achievement": "Our team, #Hawks, secured the first prize in the ML Hackathon held at Vishnu Institute of Technology, Kovvada, Andhra Pradesh."
        },
        # Add more competition entries as needed
    ],
    "publications": [
        {
            "title": "College NIRF Rank Predictor using ML",
            "publication": "IJRASET",
            "date": "04/2023"
        },
        # Add more publication entries as needed
    ],
    "copyrights": [
        {
            "title": "A Novel Software for College NIRF Rank Predictor Using Machine Learning",
            "registration_number": "SW-17552/2023",
            "date": "Nov 10, 2023"
        },
        # Add more copyright entries as needed
    ]
}

# Create and print resume JSON
resume_json = create_resume(data)
print(json.dumps(resume_json, indent=4))
```
